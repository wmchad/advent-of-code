***Work out the steps to release the most pressure in 30 minutes. What is the most pressure you can release?***

````python
import numpy as np
import re
input_file  = open("data/2022-12-16-input.txt")
input_lines = input_file.readlines()
line_pattern = r'Valve ([A-Z]{2}) has flow rate=([0-9]+); tunnels? leads? to valves? ([A-Z, ]+)'
valve_info = [re.search(line_pattern, line).groups() for line in input_lines]
valves = {valve[0]: {'flow': int(valve[1]), 'leads_to': valve[2].split(', ')}
    for valve in valve_info}
{key:value for (key, value) in valves.items() if value['flow'] > 0}
````

````python
def DistanceBetween(from_valve, to_valve, valves, visited_valves):
    if from_valve == to_valve:
        return 0
    else:
        return 1 + min([DistanceBetween(next_valve, to_valve, valves, visited_valves + [from_valve])
                        for next_valve in valves[from_valve]['leads_to']
                        if next_valve not in visited_valves],
                      default = len(valves) + 1)
    
DistanceBetween('AA', 'JT', valves, [])
````

````python
interesting_valves = ['AA'] + [key for (key, value) in valves.items() if value['flow'] > 0]
distances = np.array([[DistanceBetween(interesting_valves[i], interesting_valves[j], valves, [])
  for i in range(len(interesting_valves))]
 for j in range(len(interesting_valves))
])
flows = [0] + [value['flow'] for value in valves.values() if value['flow'] > 0]
distances
````

````python
minutes_left = 30

def MaxPressureRelease(cur_valve, minutes_left, distances, flows, used_valves):
    if minutes_left <= 1:
        return 0
    cur_release = 0
    if cur_valve not in used_valves and flows[cur_valve] > 0:
        minutes_left -= 1
        cur_release = flows[cur_valve] * minutes_left
    max_release = 0
    for next_valve in range(len(flows)):
        if next_valve not in used_valves:
            max_release = max(
                max_release,
                MaxPressureRelease(next_valve,
                                   minutes_left - distances[cur_valve, next_valve],
                                   distances,
                                   flows,
                                   used_valves + [cur_valve])
            )
    return cur_release + max_release

MaxPressureRelease(0, minutes_left, distances, flows, [0])
````

***With you and an elephant working together for 26 minutes, what is the most pressure you could release?***

````python
minutes_left = 26

def TwoActorMaxPressureRelease(
    actor1_valve, actor1_distance,
    actor2_valve, actor2_distance,
    minutes_left, distances, flows, remaining_valves):
    if minutes_left <= 1:
        return 0
    minutes_left -= 1
    match (actor1_distance, actor2_distance):
        case (0, 0):
            #turn on both valves
            cur_release = flows[actor1_valve] * minutes_left + flows[actor2_valve] * minutes_left
            #select next destination for both actors
            max_release = 0
            for next_valve1 in remaining_valves:
                next_distance1 = distances[actor1_valve, next_valve1]
                new_remaining_valves = remaining_valves.copy()
                new_remaining_valves.remove(next_valve1)
                max_release = max(max_release, flows[next_valve1] * (minutes_left - next_distance1 - 1))
                for next_valve2 in new_remaining_valves:
                    next_distance2 = distances[actor2_valve, next_valve2]
                    final_remaining_valves = new_remaining_valves.copy()
                    final_remaining_valves.remove(next_valve2)
                    max_release = max(
                        max_release,
                        TwoActorMaxPressureRelease(
                            next_valve1, next_distance1,
                            next_valve2, next_distance2,
                            minutes_left, distances, flows, final_remaining_valves)
                    )
            return cur_release + max_release
        case (0, d):
            #turn on first valve
            cur_release = flows[actor1_valve] * minutes_left
            #reduce distance for actor2
            #select next destination for actor1
            max_release = max(0, flows[actor2_valve] * (minutes_left - actor2_distance))
            for next_valve in remaining_valves:
                new_remaining_valves = remaining_valves.copy()
                new_remaining_valves.remove(next_valve)
                max_release = max(
                    max_release,
                    TwoActorMaxPressureRelease(
                        next_valve, distances[actor1_valve, next_valve],
                        actor2_valve, actor2_distance - 1,
                        minutes_left, distances, flows, new_remaining_valves)
                )
            return cur_release + max_release
        case (d, 0):
            # refund the time and switch actor positions
            return TwoActorMaxPressureRelease(
                actor2_valve, actor2_distance,
                actor1_valve, actor1_distance,
                minutes_left + 1, distances, flows, remaining_valves)
        case (d1, d2):
            #reduce distance for actor1
            #reduce distance for actor2
            return TwoActorMaxPressureRelease(
                actor1_valve, actor1_distance - 1,
                actor2_valve, actor2_distance - 1,
                minutes_left, distances, flows, remaining_valves)

TwoActorMaxPressureRelease(0, 0, 0, 0, minutes_left + 1, distances, flows, list(range(len(flows))))
# 2519 too low
````

````python
def GreedyNextValve(remaining_valves, distances, flows, time_left):
    #print((time_left - distances - 1) * flows)
    best_valve = 0
    best_flow = 0
    for i in range(len(remaining_valves)):
        valve_i = remaining_valves[i]
        valve_flow = (time_left - distances[valve_i] - 1) * flows[valve_i]
        if valve_flow > best_flow:
            best_flow = valve_flow
            best_valve = i
    return remaining_valves[best_valve]

def TwoActorMaxPressureRelease(
    actor1_valve, actor1_distance,
    actor2_valve, actor2_distance,
    minutes_left, distances, flows, remaining_valves):
    if minutes_left <= 1:
        return 0
    minutes_left -= 1
    next_remaining_valves = remaining_valves.copy()
    next_valve1 = actor1_valve
    next_distance1 = actor1_distance - 1
    next_valve2 = actor2_valve
    next_distance2 = actor2_distance - 1
    cur_release = 0
    if actor1_distance == 0:
        cur_release += flows[actor1_valve] * minutes_left
        next_valve1 = GreedyNextValve(next_remaining_valves, distances[actor1_valve, :], flows, minutes_left)
        print(f'{minutes_left} minutes left')
        print(f'actor 1 going to valve {next_valve1}')
        next_remaining_valves.remove(next_valve1)
        next_distance1 = distances[actor1_valve, next_valve1]
    if actor2_distance == 0:
        cur_release += flows[actor2_valve] * minutes_left
        next_valve2 = GreedyNextValve(next_remaining_valves, distances[actor2_valve, :], flows, minutes_left)
        print(f'{minutes_left} minutes left')
        print(f'actor 2 going to valve {next_valve2}')
        next_remaining_valves.remove(next_valve2)
        next_distance2 = distances[actor2_valve, next_valve2]
    return cur_release + TwoActorMaxPressureRelease(
        next_valve1, next_distance1,
        next_valve2, next_distance2,
        minutes_left, distances, flows, next_remaining_valves)

minutes_left = 26
TwoActorMaxPressureRelease(0, 0, 0, 0, minutes_left + 1, distances, flows, list(range(len(flows))))
````
