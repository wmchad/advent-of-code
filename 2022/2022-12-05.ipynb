After the rearrangement procedure completes, what crate ends up on top of each stack?

````python
input_file = open("data/2022-12-05-input.txt")
input_lines = input_file.readlines()

stacks = [[] for _ in range(9)]
for line in input_lines:
    if '[' in line:
        crates = line[1::4]
        for i in range(9):
            if crates[i] != ' ': stacks[i].insert(0, crates[i])
    else:
        match line.strip().split(' '):
            case [_, n, _, from_stack, _, to_stack]:
                for niter in range(int(n)):
                    stacks[int(to_stack) - 1].append(stacks[int(from_stack) - 1].pop())
print('The top crates are ' + ''.join([stack[-1] for stack in stacks]))
````
