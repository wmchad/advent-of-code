As you go to input this number on Bob's console, you discover that some buttons you expected are missing. Instead, you are met with buttons labeled =, -, 0, 1, and 2. Bob needs the input value expressed as a SNAFU number, not in decimal.

Reversing the process, you can determine that for the decimal number 4890, the SNAFU number you need to supply to Bob's console is 2=-1=0.

````python
def ConvertSnafuToNumber(snafu):
    total = 0
    for i in range(len(snafu)):
        match snafu[len(snafu) - 1 - i]:
            case '=':
                total += -2 * (5 ** i)
            case '-':
                total += -1 * (5 ** i)
            case '1':
                total += 5 ** i
            case '2':
                total += 2 * (5 ** i)
    return total
````

````python
import numpy as np

snafu_dict = {
    '0': '0',
    '1': '1',
    '2': '2',
    '-1': '-',
    '-2': '='
}

def ConvertNumberToSnafu(number):
    snafu = ''
    power = 0
    last_digit = 0
    current_sum = 0
    while True:
        if 2 * (5 ** (power - 1)) + 1 > number:
            break
        last_digit = int(((number - current_sum) // (5 ** power)) % 5)
        last_digit = last_digit - 5 * (last_digit // 3)
        current_sum += last_digit * (5 ** power)
        snafu = snafu_dict[str(last_digit)] + snafu
        power += 1
    return snafu
````

````python
input_file  = open("data/2022-12-25-input.txt")
input_lines = input_file.readlines()
fuel_total  = sum([ConvertSnafuToNumber(line.strip()) for line in input_lines])
print(f'The SNAFU number to provide is {ConvertNumberToSnafu(fuel_total)}')
````
