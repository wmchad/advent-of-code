With everything replanted for next year (and with elephants and monkeys to tend the grove), you and the Elves leave for the extraction point.

Partway up the mountain that shields the grove is a flat, open area that serves as the extraction point. It's a bit of a climb, but nothing the expedition can't handle.

At least, that would normally be true; now that the mountain is covered in snow, things have become more difficult than the Elves are used to.

As the expedition reaches a valley that must be traversed to reach the extraction site, you find that strong, turbulent winds are pushing small blizzards of snow and sharp ice around the valley. It's a good thing everyone packed warm clothes! To make it across safely, you'll need to find a way to avoid them.

Fortunately, it's easy to see all of this from the entrance to the valley, so you make a map of the valley and the blizzards (your puzzle input). For example:


The walls of the valley are drawn as `#`; everything else is ground. Clear ground - where there is currently no blizzard - is drawn as `.`. Otherwise, blizzards are drawn with an arrow indicating their direction of motion: up (`^`), down (`v`), left (`<`), or right (`>`).

The above map includes two blizzards, one moving right (`>`) and one moving down (`v`). In one minute, each blizzard moves one position in the direction it is pointing:

***What is the fewest number of minutes required to avoid the blizzards and reach the goal?***

````python
dir_stay  = [0, 0]
dir_down  = [1, 0]
dir_up    = [-1, 0]
dir_right = [0, 1]
dir_left  = [0, -1]
possible_steps = [dir_stay, dir_down, dir_up, dir_right, dir_left]

def GetBlizzards(valley_map):
    blizzards = []
    directions = []
    for i in range(valley_map.shape[0]):
        for j in range(valley_map.shape[1]):
            match valley_map[i, j]:
                case 'v':
                    blizzards.append([i, j])
                    directions.append(dir_down)
                case '^':
                    blizzards.append([i, j])
                    directions.append(dir_up)
                case '>':
                    blizzards.append([i, j])
                    directions.append(dir_right)
                case '<':
                    blizzards.append([i, j])
                    directions.append(dir_left)
    return (blizzards, directions)

def MoveBlizzard(blizzard, direction, map_shape):
    new_blizzard = [blizzard[0] + direction[0], blizzard[1] + direction[1]]
    if new_blizzard == [-1, 0] or new_blizzard == [map_shape[0], map_shape[1] - 1]:
        return new_blizzard
    if new_blizzard == [-2, 0]:
        return [map_shape[0] - 1, 0]
    if new_blizzard == [map_shape[0] + 1, map_shape[1] - 1]:
        return [0, map_shape[1] - 1]
    return [new_blizzard[0] % map_shape[0], new_blizzard[1] % map_shape[1]]

def MoveBlizzards_old(blizzards, directions, map_shape):
    return [
        [(blizzards[i][0] + directions[i][0]) % map_shape[0],
         (blizzards[i][1] + directions[i][1]) % map_shape[1]]
        for i in range(len(blizzards))
    ]

def MoveBlizzards(blizzards, directions, map_shape):
    return [
        MoveBlizzard(blizzards[i], directions[i], map_shape)
        for i in range(len(blizzards))
    ]

def ConvertDirectionToArrow(direction, count):
    if count > 1:
        return str(count)
    match direction:
        case x if x == dir_down:
            return 'v'
        case x if x == dir_up:
            return '^'
        case x if x == dir_right:
            return '>'
        case x if x == dir_left:
            return '<'
    
def PrintValleyRow(blizzards, directions, row_index, width):
    print('#' + ''.join(
        [
            ConvertDirectionToArrow(directions[blizzards.index([row_index, i])],
                                    blizzards.count([row_index, i]))
            if [row_index, i] in blizzards else '.'
            for i in range(width)
        ]
    ) + '#')
    
def PrintValley(blizzards, directions, map_shape):
    print('#.' + '#' * (map_shape[0] + 2))
    for i in range(map_shape[0]):
        PrintValleyRow(blizzards, directions, i, map_shape[1])
    print('#' * (map_shape[0] + 2) + '.#')
    
def LocationInMap(location, map_shape):
    if location == [-1, 0]:
        return True
    if location == [map_shape[0], map_shape[1] - 1]:
        return True
    if location[0] < 0 or location[0] >= map_shape[0] or location[1] < 0 or location[1] >= map_shape[1]:
        return False
    return True
    
def PossibleMoves(current_location, blizzards, map_shape):
    possible_moves = []
    for move in possible_steps:
        new_location = [current_location[0] + move[0], current_location[1] + move[1]]
        if new_location not in blizzards and LocationInMap(new_location, map_shape):
            possible_moves.append(new_location)
    return possible_moves

def StepsToLocation(current_locations, target_location, blizzards, directions, map_shape):
    if target_location in current_locations:
        return (0, blizzards)
    next_blizzards = MoveBlizzards(blizzards, directions, map_shape)
    next_locations = []
    for cur_loc in current_locations:
        for move in PossibleMoves(cur_loc, next_blizzards, map_shape):
            if move not in next_locations:
                next_locations.append(move)
    (steps, final_blizzards) = StepsToLocation(next_locations, target_location, 
                                               next_blizzards, directions, map_shape)
    return (steps + 1, final_blizzards)
````

As the expedition reaches the far side of the valley, one of the Elves looks especially dismayed:

Since you're so good at dodging blizzards, the Elves humbly request that you go back for his snacks. From the same initial conditions, how quickly can you make it from the start to the goal, then back to the start, then back to the goal?

In the above example, the first trip to the goal takes 18 minutes, the trip back to the start takes 23 minutes, and the trip back to the goal again takes 13 minutes, for a total time of 54 minutes.

***What is the fewest number of minutes required to reach the goal, go back to the start, then reach the goal again?***

````python
def StepsToMultipleLocations(start_location, target_locations, blizzards, directions, map_shape):
    current_start_location = start_location
    current_blizzards = blizzards
    total_steps = 0
    for current_target in target_locations:
        current_steps, current_blizzards = StepsToLocation(
            [current_start_location], current_target,
            current_blizzards, directions, map_shape
        )
        print(current_steps)
        current_start_location = current_target
        total_steps += current_steps
    return total_steps
````
